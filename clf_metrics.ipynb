```python
import numpy as np
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
```

```python
class0_len = 1000
x0 = np.random.normal((0,1,2),(1,2,2),(class0_len,3))
y0 = [0]*class0_len

class1_len = 2000
x1 = np.random.normal((1,1,-1),(3,1,1),(class1_len,3))
y1 = [1]*class1_len
```

```python
print(x1)
y1
```

```python
X = np.concatenate((x0,x1))
y = y0 +y1
```

```python
X_train , X_val , y_train, y_val = train_test_split(X,y,random_state = 24)
```

```python
clf = LogisticRegression()
clf.fit(X_train,y_train)
print(clf.score(X_train,y_train))
print(clf.score(X_val,y_val))
```

```python
y_val
```

```python
def perf_measure(y_actual,y_hat):
    
    tp = 0
    tn = 0
    fp = 0
    fn = 0
    
    for i in range(len(y_hat)):
        if y_actual[i]==y_hat[i] and y_hat[i]==1:
            tp = tp + 1
        if y_actual[i]==y_hat[i] and y_hat[i]== 0:
            tn = tn + 1
        if y_actual[i]!=y_hat[i] and y_hat[i]==1:
            fp = fp + 1
        if y_actual[i]!=y_hat[i] and y_hat[i]==0:
            fn = fn + 1
        
    
    return (tp,fp,tn,fn)
```

```python
tp,fp,tn,fn = perf_measure(y_val, clf.predict(X_val))
```

```python
def all_scores(tp,fp,tn,fn):
    precision = tp/(tp+fp)
    recall = tp/(tp+fn)
    specificity = tn/(tn+fp)
    npv = tn/(tn+fn)
    
    f1_score = 2*precision*recall/(precision+recall)
    accuracy = (tp+tn)/(tp+fp+tn+fn)
       
    print("precision:",precision)
    print("recall:",recall)
    print("specificity:",specificity)
    print("npv:",npv)
    
    print("f1_score",f1_score)
    print("accuracy:",accuracy)
```

```python
all_scores(620,1,50,150)
```
